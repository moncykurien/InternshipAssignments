#2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```
import pandas as pd
```

```
pd.set_option('display.max_rows', 100)
```

```
def get_top_100(urls, file_name, region = None):
  movie_name_list = []
  release_year_list = []
  rating_list = []

  for url in urls:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    for movie in soup.find_all('h3'):
      try:
        movie_name_list.append(movie.text.split('\n')[2].strip())
        release_year_list.append(movie.text.split('\n')[3].split('(')[-1].strip(')').strip())
      except:
        continue
    if region == None:
      cls = "inline-block ratings-imdb-rating"
    elif region.casefold() == 'india':
      cls = "ipl-rating-star small"
    else:
      raise Exception("The 'region' parameter should be 'India' or None.")
    for rating in soup.find_all('div', class_ = cls):
      try:
        rating_list.append(rating.text.split()[0])
      except:
        continue
  movie_name_list=list(filter(None,movie_name_list))
  release_year_list=list(filter(None,release_year_list))
  rating_list=list(filter(None,rating_list))

  dataframe = pd.DataFrame({'Name':movie_name_list,'IMDB rating':rating_list,'Year of release': release_year_list})
  dataframe.to_csv('/content/'+file_name)
  return dataframe
```

```
urls = ['https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc','https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt']
top100 = get_top_100(urls, 'IMDB_top100_movies.csv')
top100.head(100)
```

#5. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:
i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. 

ii) Top 10 ODI Batsmen in men along with the records of their team and rating. 

iii) Top 10 ODI bowlers along with the records of their team and rating.

```
def get_teams(url, limit):

  country_list = []
  matches_list = []
  points_list = []
  ratings_list = []
  match_point_temp_list=[]  
  focus_page = requests.get(url) 
  focus_soup = BeautifulSoup(focus_page.content, 'html.parser')
  rows = focus_soup.find_all('tr', class_ = 'rankings-block__banner')
  for row in rows:
    country_list.append(row.find('span',class_='u-hide-phablet').text.strip())
    matches_list.append(row.find('td',class_='rankings-block__banner--matches').text.strip())
    points_list.append(row.find('td',class_='rankings-block__banner--points').text.strip())
    ratings_list.append(row.find('td',class_='rankings-block__banner--rating u-text-right').text.strip())

  rows = focus_soup.find_all('tr', class_ = 'table-body', limit=limit-1)
  for row in rows:
    for country in row.findChildren('span',class_='u-hide-phablet'):
      country_list.append(country.text.strip())
    for matches in row.findChildren('td',class_='table-body__cell u-center-text'):
      match_point_temp_list.append(matches.text)
    for rating in row.findChildren('td',class_='table-body__cell u-text-right rating'):
      ratings_list.append(rating.text.strip())
  matches_list.extend(match_point_temp_list[:-1:2])
  points_list.extend(match_point_temp_list[1::2])
  return pd.DataFrame({'Team': country_list, 'Matches':matches_list, 'Points':points_list, 'Ratings':ratings_list})


def get_player(url, limit):

  players_list = []
  countries_list = []
  pl_ratings_list = []
  focus_page = requests.get(url) 
  focus_soup = BeautifulSoup(focus_page.content, 'html.parser')
  players_list.append(focus_soup.find('div',class_='rankings-block__banner--name-large').text.strip())
  countries_list.append(focus_soup.find('div',class_='rankings-block__banner--nationality').text.strip())
  pl_ratings_list.append(focus_soup.find('div',class_='rankings-block__banner--rating').text.strip())

  names = focus_soup.find_all('td', class_ = 'table-body__cell rankings-table__name name',limit=limit-1)
  teams = focus_soup.find_all('span', class_ = 'table-body__logo-text',limit=limit-1)
  ratings = focus_soup.find_all('td', class_ = 'table-body__cell rating',limit=limit-1)

  for name, team, rating in zip(names, teams, ratings):
    players_list.append(name.text.strip())
    countries_list.append(team.text.strip())
    pl_ratings_list.append(rating.text.strip())
  names, teams, ratings = (None, None, None)
  return pd.DataFrame({'Player': players_list, 'Team': countries_list, 'Rating': pl_ratings_list})


def scrap_cricket(url, category, match_type, limit, player_role=None, gender='men'):
  focus_url = None
  page = requests.get(url)
  soup = BeautifulSoup(page.content, 'html.parser')
  if category == 'teams':
    print(f"The top {limit} teams are :")
    focus_url = url+soup.find('a', text="Team Rankings")['href']+'/'+match_type
    if gender == 'women':
      focus_url=focus_url.replace('mens','womens')
    return get_teams(focus_url, limit)
  elif category == 'players':
    print(f"The top {limit} players in {player_role} are :")
    focus_url = url+soup.find('a', text="Players Rankings")['href']+'/'+match_type+'/'+player_role
    if gender == 'women':
      focus_url=focus_url.replace('mens','womens')
    print(focus_url)
    return get_player(focus_url, limit)

```

#7. Write a python program to scrape details of all the mobile phones under Rs.20,000 listed on Amazon.in. The scraped data should include Product Name, Price, Image URL and Average Rating.

```
url = 'https://www.amazon.in'
requests.get(url)
```

```
def get_amazon_products(url):
  user_win = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
  user_linux = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36'
  HEADERS = ({'User-Agent': user_win,
              'Accept-Language': 'en-US, en;q=0.5'})
 

  titles = []
  prices = []
  images = []
  ratings = []
  base_url = 'https://www.amazon.in'
  url = url
  has_url = True
  page = 1

  while has_url:
    print(f"Response for page number {page} : {requests.get(url)}")
    soup = BeautifulSoup(requests.get(url, headers= HEADERS).content, 'html.parser')
    titles_soup = soup.find_all('a',class_='a-link-normal a-text-normal')
    prices_soup = soup.find_all('span','a-price-whole')
    images_soup = soup.find_all('img',class_='s-image')
    ratings_soup = soup.find_all('span',class_='a-icon-alt')
    for title, price, image, rating in zip(titles_soup, prices_soup, images_soup, ratings_soup):
      titles.append(title.text.strip())
      prices.append(price.text.strip())
      images.append(image['src'])
      ratings.append(rating.text.split()[0])
      
    next = soup.find('li', class_='a-last')
    page = page + 1
    try:
      url = base_url+next.findChild('a')['href']
    except:
      has_url = False
    
  return pd.DataFrame({'Product Name':titles,'Price':prices,'Image Url':images,'Rating out of 5':ratings})
```

#8. Write a python program to extract information about the local weather from the National Weather Service website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day extended forecast display for the city. The data should include period, short description, temperature and description. 

```
def get_forcast(url):
  
  periods_list = []
  forecasts = []

  soup = BeautifulSoup(requests.get(url).content, 'lxml')
  periods = soup.find_all('div',class_='col-sm-2 forecast-label')
  forcasts_texts = soup.find_all('div',class_='col-sm-10 forecast-text')

  for period,text in zip(periods,forcasts_texts):
    periods_list.append(period.text.strip())
    forecasts.append(text.text.strip())

  initial_dataframe = pd.DataFrame({'Period':periods_list, 'Description': forecasts})
  initial_dataframe['ShortDescription'] = initial_dataframe['Description'].apply(lambda x : x.split(',')[0])
  initial_dataframe['Temperature'] = initial_dataframe['Description'].apply(lambda x : x.split(',')[1].split('.')[0].split()[-1])

  return initial_dataframe
```

```
url = 'https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YFclBK8zZPY'
get_forcast(url)
```
